```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

# dataset.columns

```python
Independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
Dependent=dataset[['charges']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Independent=sc.fit_transform(Independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'splitter':['best', 'random'],'max_features':[None,'sqrt','log2']}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(Independent,Dependent)
```

```python
re=grid.cv_results_
y_pred=grid.predict(Independent)
from sklearn.metrics import r2_score
r_score = r2_score(Dependent,y_pred)
r_score
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
import pickle
filename = "Finalized_model_DecisionTree_Regression.sav"
pickle.dump(grid,open(filename,'wb'))
```
