```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
Independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=dataset[['charges']]
Dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['rbf','poly','sigmoid','linear'],'C':[1,10,100,200,500],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
y_pred=grid.predict(x_test)
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
import pickle
filename="SVM_Preprocessed_input&output.sav"
pickle.dump(grid,open(filename,'wb'))
pre_input=sc.transform([[19,39,0,1,0]])
pre_input
```

```python
loaded_model=pickle.load(open("SVM_Preprocessed_input&output.sav",'rb'))
result=loaded_model.predict(pre_input)
result
```
